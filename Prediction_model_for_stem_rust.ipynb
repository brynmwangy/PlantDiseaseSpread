```python
cleaned_data = data_frames.drop(['Location.ID', 'HostGenusID', 'HostGenusNameScientific',
       'HostSpeciesID', 'HostSpeciesNameScientific', 'HostCultivarName',
       'UserInit', 'ObsYear', 'ObsDate', 'SurveySiteID', 'SurveySiteName',
       'SurveyorName', 'InstitutionName', 'LocationName', 'Latitude',
       'Longitude', 'Altitude', 'FieldArea', 'GrowthStageID',
       'GrowthStageName', 'Comment', 'StemRust.Severity',
       'StemRust.SeverityName', 'StemRust.Incidence',
       'StemRust.IncidenceName', 'StemRust.InfectionType',
       'StemRust.InfectionTypeName', 'LeafRust.Severity',
       'LeafRust.SeverityName', 'LeafRust.Incidence',
       'LeafRust.IncidenceName', 'LeafRust.InfectionType',
       'LeafRust.InfectionTypeName', 'YellowRust.Severity',
       'YellowRust.SeverityName', 'YellowRust.Incidence',
       'YellowRust.IncidenceName', 'YellowRust.InfectionType',
       'YellowRust.InfectionTypeName','YellowRust.Binary', 'NoRust.Binary'], axis=1)

cleaned_data
```

```python
def f(x):
    result = x
    if x == True:
        result = 1
    elif x == False:
        result = 0
    return result;
f(34)
```

```python
cleaned_data2 = cleaned_data.applymap(f)
```

```python
cleaned_data2
```

```python
cleaned_data3 = cleaned_data2[np.isfinite(cleaned_data2[target_field])]
X = cleaned_data3.drop([target_field], axis=1)
X = X.fillna(X.mean(0))
Y = cleaned_data3[target_field]
print(X.shape)
print(Y.shape)
```

```python
clf = classifier(n_estimators=100)
model = clf.fit(X,Y)
```

```python
X_train, X_test, Y_train, Y_test = cross_validation.train_test_split(
...     X, Y, test_size=0.2, random_state=0)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
model = clf.fit(X_train, Y_train)
accuracy = model.score(X_test, Y_test)
```

```python
from sklearn.metrics import f1_score
Y_predicted = model.predict(X_test)
precision = precision_score(Y_test, Y_predicted, average='binary')
recall = recall_score(Y_test, Y_predicted, average='binary')
f1_score = f1_score(Y_test, Y_predicted, average='binary')
print("Accuracy=", accuracy, " Precision=", precision, "Recall=", recall, "F1 score=", f1_score)
```
